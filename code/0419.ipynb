# 88 Merge Sorted Array 
In place operation 

step
1. In place 
2. they are already sorted, only need to merge, the merge sorting part merge 
3. give the length, not the index. 
4. While inside, is the index >=0 and index2 >=0
5. If nums1 is left, not need to change , if num2 is left, then continue to append 


exp: [1， 2， 3， 0， 0， 0 ], [2, 5, 6]  
[_, _, _, 0, 0, 6 ]
[_, _, _, 0, 5, 6 ]
[_, _, _, 3, 5, 6 ]
[_, _, 2, 3, 5, 6 ]
the rest of nums1 will not change 


edge case when the first arrary is zero. the second will be merge to nums2 


```python
class Solution(object):
    
    def merge( self, nums1, m, nums2, n): 
        p = m + n -1 
        i = m - 1 
        j = n -1 
        
        while i >= 0 and j>=0 : 
            if nums1[i] >= nums2[j]: 
                nums1[p] = nums1[i]
                i -=1 
                
            else: 
                nums1[p] = nums2[j]
                j -=1 
                
            p -=1 
            
        while j >= 0: 
            nums1[p] = nums2[j]
            j -=1 
            p -=1 
            
            
        return nums1
                
            
    
    
```

```python
class Solution(object):
    def merge(self, nums1, m, nums2, n):
        p = m + n - 1   # nums1 最后一个位置
        i = m - 1       # nums1 有效部分最后一个
        j = n - 1       # nums2 最后一个

        while i >= 0 and j >= 0:
            if nums1[i] > nums2[j]:
                nums1[p] = nums1[i]
                i -= 1
            else:
                nums1[p] = nums2[j]
                j -= 1
            p -= 1

        # 如果 nums2 还有剩余，补到前面
        while j >= 0:
            nums1[p] = nums2[j]
            j -= 1
            p -= 1
```
