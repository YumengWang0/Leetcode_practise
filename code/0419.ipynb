# 53 Maximum Subarray
divide and conquer? how
finding the subarray. nums[l: r] that has the largest sum. 
任何一个最大子数组，只可能是三种情况之一：left side， right side or the crossed left to right

No sorting on the left and right. 
But the crossing : left max , and right max 

Example 1:

Input: nums = [-2,1,-3,4,-1,2,1,-5,4]
Output: 6
Explanation: The subarray [4,-1,2,1] has the largest sum 6.
Example 2:

Input: nums = [1]
Output: 1
Explanation: The subarray [1] has the largest sum 1.
Example 3:

Input: nums = [5,4,-1,7,8]
Output: 23
Explanation: The subarray [5,4,-1,7,8] has the largest sum 23.
  

```python
class Solution(object):
    # the index is really import for the merge compare , find the left sum max and the right sum max for the crossing part 
    
    def maxSubArray(self, nums):
        def helper(left, right):
            if left == right:
                return nums[left]

            mid = (left + right) // 2

            left_best = helper(left, mid)
            right_best = helper(mid + 1, right)
            cross_best = self.find_maximum(nums, left, mid, right)
            return max(left_best, right_best, cross_best)

        return helper(0, len(nums) - 1)  
    
    
    
    def find_maximum( self, nums, left, mid, right): 
         
        # Here to find the maximum crossing 
        curr = 0 
        left_sum = float('-inf') 
        
        # from the middle to the left, left -1 is not attend 
        for i in range( mid, left -1, -1 ):
            curr += nums[i]
            left_sum = max(left_sum, curr )
            
            
        # find the right side 
        curr = 0 
        right_sum = float('-inf')
        for i in range(mid-1, right+1):  
            curr += nums[i] 
            right_sum = max(right_sum, curr) 
            
        return left_sum + right_sum 
            
            
        
```
