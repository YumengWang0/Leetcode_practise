```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next

class Solution(object):
    def reverseList(self, head):
        """
        :type head: Optional[ListNode]
        :rtype: Optional[ListNode]
        """

        curr = head 
        prev = None 

        while curr is not None: 

            temp = curr.next 
            curr.next = prev 
            prev = curr 
            curr = temp 

        # return to the head 

        return  prev


# text case 

# head = [1,2,3,4,5], [1], [1, 2], [] 
```

# 21 Merge Two Sorted Lists 
the two listnode object are sorted.

list1 = [1,2,4], list2 = [1,3,4]
Output: [1,1,2,3,4,4]

use the dummy to initialize a head, return the dummy.next will remove the head node 

step:
- dummy = None is set  
- compare the head of list1 and the head of list2, the smaller to the dummy.next 
- move the pointer to the smaller head, for example. list1 = list1.next if list1 is smlller . 
- while list1 or list2 is None, then append the rest of the one has rest part 


**key** : 
- dummy should initilized with ListNode(0) 
- dummy should assign with a tail to copy it , tail will move but dummy is the whole listnode 




```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution(object):
    def mergeTwoLists(self, list1, list2):
        """
        :type list1: Optional[ListNode]
        :type list2: Optional[ListNode]
        :rtype: Optional[ListNode]
        """ 

        
        all_dummy = ListNode(0)
        dummy = all_dummy
        
        
        # compare the value of the two lists , with the list1.val and list2.val 
        while list1 is not None and list2 is not None: 
            if list1.val <= list2.val: 
                dummy.next = list1
                list1 = list1.next

            else: 
                dummy.next = list2
                list2 = list2.next 
            # this part is import 
            dummy = dummy.next 

        if list1 is not None: 
            dummy.next = list1 

        elif list2 is not None: 
            dummy.next = list2 

        return all_dummy.next 


 


```

slow fast two ppinters 

if the slow and fast pointer they meet, it means they find the  circle and reset the fast pointer to move the same pace as the slow pointer, it will reach to the same position, which is the cycle start point 

step:  This is to prove the cycle exist , not to find the position 
1. set the slow = 0 
2. set the fast = 2 * slow 
3. while slow != fast, then slow ++ and fast ++ 2 
4. For the listnode, the only difference is curr.next.next for fast and curr.next for slow, which correspoding to the slow ++ and fast ++2 

 




```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution(object):
    def hasCycle(self, head):
        """
        :type head: ListNode
        :rtype: bool
        """


        slow = head 
        fast = head



        # there must be circle if they meet 
        while fast is not None and fast.next is not None : 
        
            slow = slow.next 
            fast = fast.next.next 

            if slow == fast:
                return True

        return False  


        

        
         
```

```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution(object):
    def removeNthFromEnd(self, head, n):
        """
        :type head: Optional[ListNode]
        :type n: int
        :rtype: Optional[ListNode]
        """
        
        #set the next listnode

        dummy = ListNode(0)
        dummy.next = head 

        slow = dummy 
        fast = dummy 
        
        # stop at the position of the fast, which is n steps ahead of the slow 
        for _ in range(n):
            fast = fast.next 
            
            
        # fast is the last one 
        while fast.next is not None:
            slow = slow.next
            fast = fast.next 
            
            
        slow.next = slow.next.next 
        
    
        
        return dummy.next 
    
    
#exp: list = [1,2,3,4,5], n = 2  # output: [1,2,3,5]
#epx: list = [1], n = 1  # output: []
#exp: list = [1,2], n = 2  # output: [2] remove the head 

```
